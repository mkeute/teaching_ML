## Algorithms
Our goal is now to fit a model to this data. Similar to a linear regression model, we want it to 1.) capture the pattern of data that we see and 2.) be able to predict class membership for unseen data. There are several **classification algorithms** that can achieve this. We will now look at the most important ones.

### k nearest neighbours (kNN)
This is probably the simplest classification algorithml:
* Choose a number k (number of neighbours)
* For a new data point (i.e., BMI and BP data from a new patient with unknown gender), you predict class membership based on the majority vote from its k nearest neighbours 
* The choice of k determines the overfitting / underfitting behavior of the algorithm
* Advantage: very simple, very few assumptions
* Disadvantage: very inefficient (you need to carry the whole data along)

### Logistic Regression
Logistic Regression fits a regression model to the data, with one important modification: while in a linear regression model, we predicted the value for a new subject j $Y_j \in (-\infty, \infty)$, we now want to predict $Y_j \in \{0,1\}$. We therefore let the model output a probability $P(Y_j=1|X_j)$. We use the **logistic function** to overcome the bounded-ness of output probabilities. The logistic function will always have a value between 0 and 1, so it is suitable for modeling probabilities:

$ logistic(x) = \frac{1}{1+exp(-x)}$.

Our regression model becomes:

$P(Y_j=1) = logistic(\boldsymbol{X}\boldsymbol{\beta}) = \frac{1}{1+exp(-\boldsymbol{X}\boldsymbol{\beta})}$.

Based on the probability that our model predicts for new data, we can assign the most likely class membership.
